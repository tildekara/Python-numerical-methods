# RANDOMWALK

```python
import random
random.random()
from random import random

#bladzenie losowe
N = 1000 #liczba krokow
x, y = 0, 0 #punkt poczatkowy
x_lista, y_lista = [],[]

for i in range(N):
    x_lista.append(x)
    y_lista.append(y)
    
    x +=random()-0.5
    y +=random()-0.5
    
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x_lista, y_lista)
```

# rndomwalk 2D 1000 czastek

```python
N = 10000
Nczastek = 1000
Ndim = 2

r = np.zeros((Nczastek,Ndim))
r.shape #zwraca rozmiar macierzy

r_array = np.zeros((Nczastek, Ndim, N))

for i in range(N):
    r_array[:,:,i] = r
    r = r+np.random.random(r.shape)-0.5 #random z np daje macierz, liczbe odejmuje od kazdego elementu


```

```python
figure, axes = plt.subplots(figsize=(12,8)) #zwiększa rozmiar wykresu
for i in range(Nczastek):
    x_tej_czastki = r_array[i,0,:]
    y_tej_czastki = r_array[i,1,:]
    axes.plot(r_array[i,0,:], r_array[i,1,:] , "-", alpha =0.3) #i <- ktora cząstka, 0,1,2 <-położenie x y lub z, : - wszystkie położenia

axes.grid()
axes.set_xlabel(r"$x \pi$") #r to row string zeby miec np. pi
axes.set_ylabel(r"$y \pi$")
axes.set_title("random walk")
```

```python
N = 100000
Nczastek = 10000
Ndim = 1

r = np.zeros((Nczastek,Ndim))
r.shape #zwraca rozmiar macierzy

r_array = np.zeros((Nczastek, Ndim, N))

for i in range(N):
    r_array[:,:,i] = r
    r = r+np.random.random(r.shape)-0.5 #random z np daje macierz, liczbe odejmuje od kazdego elementu
    
figure, axes = plt.subplots(figsize=(12,8)) #zwiększa rozmiar wykresu
axes.hist(r_array[:,0,-1].flatten(), bins = 100)
axes.grid()
axes.set_xlabel(r"$x \pi$") #r to row string zeby miec np. pi
axes.set_ylabel(r"$y \pi$")
axes.set_title("random walk")
```

```python
N = 100000
Nczastek = 1
Ndim = 1

r = np.zeros((Nczastek,Ndim))
r.shape #zwraca rozmiar macierzy

r_array = np.zeros((Nczastek, Ndim, N))

for i in range(N):
    r_array[:,:,i] = r
    r = r+np.random.random(r.shape)-0.5 #random z np daje macierz, liczbe odejmuje od kazdego elementu
    
figure, axes = plt.subplots(figsize=(12,8)) #zwiększa rozmiar wykresu
axes.hist(r_array[:,0,:].flatten(), bins = 100)
axes.grid()
axes.set_xlabel(r"$x \pi$") #r to row string zeby miec np. pi
axes.set_ylabel(r"$y \pi$")
axes.set_title("random walk")
```
